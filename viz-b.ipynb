```python
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset("tips")
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
# Create scatter plots
g = sns.FacetGrid(tips, col="sex", row="smoker", margin_titles=True)
g.map(sns.scatterplot, "total_bill", "tip")
g.fig.suptitle("tip vs bill vs gender vs smoker")
```

## Conclusions:
- it seems that there is kind of relation between tip and total bill , the higher the bill the most likely the tip is higher as well , and it fits more if non smoker males.
- The category that has lowest bill and so that tip , is non-smoker females.
- The weirdest behavior in terms of relations is the males smokers!
- Dinner tips are more than lunch tips both in numbers and values!
